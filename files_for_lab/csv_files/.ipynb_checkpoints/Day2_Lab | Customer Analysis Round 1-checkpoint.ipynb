- Perform the data cleaning operations mentioned so far in class

```python
#Delete the column education and the number of open complaints from the dataframe.
merged_files = merged_files.drop(['EDUCATION','NUMBER OF OPEN COMPLAINTS'], axis=1)
merged_files.shape
```

```python
#Correct the values in the column customer lifetime value.
#They are given as a percent, so multiply them by 100 and change `dtype` to `numerical` type.


merged_files['CUSTOMER LIFETIME VALUE_CLEANED'] = merged_files['CUSTOMER LIFETIME VALUE'].str.rstrip('%').astype('float') / 100.0
merged_files.head(10)
```

```python
merged_files.shape
```

```python
#Check for duplicate rows in the data and remove if any.
merged_files = merged_files.drop_duplicates()
merged_files.reset_index(drop=True)
```

```python
merged_files.shape
```

```python
# Filter out the data for customers who have an income of 0 or less.
merged_files[merged_files['INCOME'] >= 0]
```
